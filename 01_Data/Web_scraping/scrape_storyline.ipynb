```python
import urllib2
from bs4 import BeautifulSoup
import numpy as np
import pandas as pd
import re
```

```python
imdb = pd.read_csv("../movie_metadata.csv")
```

```python
imdb.head()
```

```python
def storyline_scraper(link):
    html = urllib2.urlopen(link)
    soup = BeautifulSoup(html, "lxml")
    try:
        desc = soup.findAll("div", {"class":"inline canwrap"})[0].find('p').text
        desc_text = re.split(r'\s{2,}', desc)[0].strip()
    except IndexError:
        desc_text = np.nan
    return desc_text
```

```python
storyline_scraper(imdb['movie_imdb_link'][3])
```

```python
imdb['storyline'] = imdb['movie_imdb_link'].apply(lambda x: storyline_scraper(x))
```

```python
imdb.to_csv("../../01_Data/imdb_with_storyline.csv", index=False)
```
